```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
%matplotlib inline
import seaborn as sns
```

```python
i = pd.read_excel('C:/Users/mjc341/desktop/Plotly/PY_MoreThan10.xlsx', dtype =object)
```

```python
ordered_i = i.sort_values(by = 'Frequency')
ordered_i = ordered_i[:20]
my_range = range(1,len(ordered_i)+1)
```

```python
my_range
```

```python
f, ax = plt.subplots(figsize = (10,8))
sns.set_style('darkgrid')
sns.set_context('talk')
plt.hlines(y = my_range, xmin =0,  xmax=ordered_i['Frequency'], color='thistle', linewidth = 5 )
plt.yticks(my_range, ordered_i['Intersection'], size = 15)
plt.title("Intersections with more than 10 Accidents", loc='center', size = 20)
plt.xlabel('Count of Incidents', size = 20)
plt.ylabel('Intersection', size = 20)
srt = plt.plot(ordered_i['Frequency'], my_range, 'o',markersize = 20, color = 'indigo',markeredgecolor = 'floralwhite', markeredgewidth = 1)
srt
```
