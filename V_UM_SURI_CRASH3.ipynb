```python
g2 = sns.PairGrid(crash1,
                  x_vars = ['total', '<15', '15-29', '30+'], y_vars = ['neighborhood'])
g2 = g2.map(plt.scatter)
```

```python
with  sns.plotting_context( 'notebook' ,font_scale = 1.5):
    g3 = sns.PairGrid(crash1, 
                  x_vars = ['total', '<15', '15-29', '30+'], y_vars = ['neighborhood'],size=10, aspect=.25)
g3.map(sns.stripplot, size=13, orient="h",
      palette="Reds_r", edgecolor="b")
g3.set(xlim=(0, 65), xlabel="", ylabel="")
# Use semantically meaningful titles for the columns
titles = ["Total Crashes", "<15 mph", "15-29 mph", "over 30 mph"]

for ax, title in zip(g3.axes.flat, titles):

    # Set a different title for each axes
    ax.set(title=title)

    # Make the grid horizontal instead of vertical
    ax.xaxis.grid(False)
    ax.yaxis.grid(True)
    
sns.despine(left=True, bottom=True)
```

```python
g3.savefig('CRASH3.png')
```
