```python
#statistical estimation "mean"
#showing average veh speed by day of week w error bars
f, ax = plt.subplots(figsize = (10,8))
plt.title("Average Vehicle Speed by Day of Week", loc='center', size = 15)
plt.xlabel('', size = 20)
plt.ylabel('', size = 20)
m = sns.barplot(x='DAYOFWEEK', y = "VEH_SPEED", data = VZ)
# replaced 0 values with the average of that column
```

```python
f, ax = plt.subplots(figsize = (10,8))
sns.set_style('darkgrid')
plt.title("Average Age of Drivers by Gender Across Years", loc='center', size = 15)
plt.xlabel('', size = 20)
plt.ylabel('', size = 20)
pp = sns.pointplot(x = 'CALYEAR', y = 'DR_AGE', hue = 'DR_GENDER', data = VZ,
                  palette = {'Male': 'g', 'Female' :'m', 'n.a.': 'y'},
                   markers = ['o','^','o'], linestyles = ['-','--','-']
                  );
#average age by gender change over year 
```
