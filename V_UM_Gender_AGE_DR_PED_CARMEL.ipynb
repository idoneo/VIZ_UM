```python
import pandas as pd
```

```python
from matplotlib import pyplot as plt
%matplotlib inline
```

```python
import seaborn as sns
```

```python
df = pd.read_csv(r'C:\Users\cdj49\Desktop\VizUM\2011-2018_Crash_Data.csv')
```

```python
df.head()
```

```python
sns.set_style('whitegrid')
```

```python
sns.violinplot(x='DR_GENDER',y='DR_AGE',data=df)

plt.title('Driver Age by Gender')
```

```python
sns.set_style('whitegrid')
```

```python
sns.violinplot(x='Ped_GENDER',y='Ped_AGE',data=df)

plt.title('Pedestrian Age by Gender')
```

```python
plt.figure(figsize=(10,6))
```

```python
sns.violinplot(x='DR_GENDER',
               y='DR_AGE',
               data=df,
              inner=None,
              palette="Blues")

sns.violinplot(x='Ped_GENDER',
              y='Ped_AGE',
             data=df,
             aplha=0.7)

plt.title('Age by Gender')
```
