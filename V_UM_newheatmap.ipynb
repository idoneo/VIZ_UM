```python
import folium as f
import pandas as pd
```

```python
from folium import plugins
from folium.plugins import HeatMap
```

```python
heat = pd.read_excel('C:/Users/mjc341/desktop/Plotly/PY_2011-18_CRASH.xlsx', dtype =object)
```

```python
heat_m = heat[heat['VEH_SPEED'] != "unknown"] 
#heat_df = heat_df[heat_df['Year']=='2007'] # Reducing data size so it runs faster
heat_m = heat[['LATITUDE', 'LONGITUDE']]
heat_m = heat.dropna(axis=0, subset=['LATITUDE', 'LONGITUDE'])
```

```python
heat_data = [[row['LATITUDE'],row['LONGITUDE']] for index, row in heat_m.iterrows()]
```

```python
miami_center = (25.761681,-80.191788)
mh = f.Map(location=miami_center, tiles='OpenStreetMap',control_scale = True, zoom_start=11)
colormap = {0.0 :'green', 0.3:'green',0.5:'darkblue', 0.7: 'cyan', 1.0 :'blue'}
mh.add_child(plugins.HeatMap(heat_data, gradient=colormap))
mh               
#HeatMap(heat_data).add_to(mh)
```
