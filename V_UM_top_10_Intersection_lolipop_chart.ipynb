```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
%matplotlib inline
import seaborn as sns
```

```python
i = pd.read_excel('C:/Users/mjc341/desktop/Plotly/PY_2011-18_by_Intersection.xlsx', dtype =object)
```

```python
i.head(3)
```

```python
ordered_i = i.sort_values(by = 'Frequency of Accidents', ascending =False)
ordered_i = ordered_i[:10]
my_range = range(1,len(ordered_i)+1)
```

```python
my_range
```

```python
f, ax = plt.subplots(figsize = (8,10))
import seaborn as sns
sns.set_style('darkgrid')
plt.hlines(y = my_range, xmin =10,  xmax=ordered_i['Frequency of Accidents'], color='skyblue', linewidth = 6 )
plt.yticks(my_range, ordered_i['Intersection'], size = 15)
plt.title("Top 10 Intersections for Pedestrian and Vehicle Accidents", loc='center', size = 20)
plt.xlabel('Count of Incidents', size = 20)
plt.ylabel('Intersection', size = 20)
srt = plt.plot(ordered_i['Frequency of Accidents'], my_range, 'o',markersize = 20, markeredgecolor = 'skyblue', markeredgewidth = 1)
srt
```
