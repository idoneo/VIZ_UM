```python
df2 = df[['Ped_GENDER','Ped_AGE']]
```

```python
df2['Ped_AGE'] = df2['Ped_AGE'].astype(str).astype(int)

df2.info()
```

```python
sns.set_style('darkgrid')
sns.set_context('talk')
f,ax =plt.subplots(figsize = (10,8))
sns.violinplot(x='Ped_GENDER',y='Ped_AGE',data=df2, palette = 'Purples_r')
plt.title('Pedestrian Age by Gender')
```

```python
sns.set_style('darkgrid')
sns.set_context('talk')
f,ax =plt.subplots(figsize = (10,8))
#ax.set_ylabel("Age")
sns.violinplot(x='Ped_GENDER' ,y='Ped_AGE',data=df2, palette = 'rainbow')
plt.xlabel('Gender')
plt.ylabel ('Age')
plt.title('Pedestrian Age by Gender')
```

```python
df['DR_AGE'] = df['DR_AGE'].astype(str).astype(int)
df['VEH_SPEED'] = df['VEH_SPEED'].astype(str).astype(int)
df['Ped_AGE'] = df['Ped_AGE'].astype(str).astype(int)
```

```python
sns.set_style('darkgrid')
sns.set_context('talk')
f,ax =plt.subplots(figsize = (10,8))
sns.violinplot(x='DR_GENDER',y='DR_AGE',data=df, palette = 'cool')
plt.xlabel('Gender')
plt.ylabel ('Age')
plt.title('Driver Age by Gender')
```

```python
sns.set_style('darkgrid')
sns.set_context('talk')
f,ax =plt.subplots(figsize = (10,8))
sns.violinplot(x='DR_GENDER',y='VEH_SPEED',data=df, palette = 'cool')
plt.xlabel('Gender')
plt.ylabel ('Speed')
plt.title('Driver Speed by Gender')
```

```python
sns.set_style('darkgrid')
sns.set_context('talk')
f,ax =plt.subplots(figsize = (10,8))
sns.violinplot(x='DAYOFWEEK',y='VEH_SPEED',data=df, palette = 'cool')
plt.ylabel ('Speed')
plt.xlabel(' ')
plt.title('Driver Speed by Day of Week')
```

```python
sns.set_style('darkgrid')
sns.set_context('talk')
f,ax =plt.subplots(figsize = (10,8))
sns.violinplot(x='Ped_GENDER',
               y='Ped_AGE',
               data=df,
              inner=None,
              palette="rainbow")

sns.violinplot(x='DR_GENDER',
              y='DR_AGE',
             data=df,palette ="cool",
             aplha=0.7)
plt.xlabel('Gender')
plt.ylabel ('Drivers vs Pedestrian Ages')
plt.title('Driver vs Pedestrian Age Comparison')

#there are more pedestrians than drivers so the back violin is pedestrians, front= driver
```

```python
sns.set_style('darkgrid')
sns.set_context('talk')
f,ax =plt.subplots(figsize = (10,8))
sns.violinplot(x='Ped_GENDER',
               y='Ped_AGE',
               data=df,
              inner=None,
              palette="cubehelix_r")
#icefire, icefire_r,coolwarm, coolwarm_r,cool,cool_r,viridis, viridis_r,spring, spring_r, summer, summer_r,cubehelix, cubehelix_r

sns.violinplot(x='DR_GENDER',
              y='DR_AGE',
             data=df,palette="cubehelix",
             aplha=0.7)

plt.title('Drivers vs Pedestrians Age and Gender')
```
