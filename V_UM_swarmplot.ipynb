```python
#Gender and VEH Speed
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
AGS = pd.read_excel('C:/Users/mjc341/desktop/Plotly/AGE_GNDR_SPD.xlsx', dtype =object)
```

```python
AGS.columns
```

```python
GS = AGS[['DR_GENDER', 'VEH_SPEED']]
```

```python
GS.head(5)
```

```python
GS.columns
```

```python
GS1 = pd.melt(GS,"DR_GENDER", var_name = 'Speed of Vehicle')
```

```python
GS1.head(5)
```

```python
f, ax = plt.subplots()
plt.rcParams['figure.figsize']=(10,10)
sns.despine(bottom = True, left = True)
```

```python
#each observation as scatterplot

sns.stripplot(x ='value', y = 'Speed of Vehicle',hue = 'DR_GENDER', data = GS1,  dodge = True,jitter = True, 
            alpha = .25, zorder = 1)
```

```python
# Conditional Means
sns.pointplot(x="value", y="Speed of Vehicle", hue="DR_GENDER",
              data=GS1, dodge=.532, join=False, palette="dark",
              markers="d", scale=.75, ci=None)
```

```python
sns.set_style('darkgrid')
with sns.plotting_context( 'notebook' ,font_scale = 1.5):
    g =sns.stripplot(x ='value', y = 'Speed of Vehicle',hue = 'DR_GENDER', data = GS1,  dodge = True,jitter = .3, 
            alpha = .25, zorder = 1, size = 8, edgecolor = 'black', linewidth = 1,)

    g1 =sns.pointplot(x="value", y="Speed of Vehicle", hue="DR_GENDER",
              data=GS1, dodge=.532, join=False, palette="dark",
              markers="s",  linestyles = '-',scale=1.5, ci=None)
    
```
